
## Exercice 2 : maximum de vraisemblance

### <span style="color:blue">QUESTION 2.1 :</span> 

Tirer 10 fois une pièce à pile ou face et modéliser les résultats obtenus comme ceux d'une variable aléatoire X qui vaut X_i=0 si on a pile et X_i=1 si on a face.

Calculez le maximum de vraisemblance du paramètre p d'un loi de Bernoulli qui modéliserait le problème. Pour y arriver, différentes valeurs possibles de p seront testées et le p retenu sera celui qui a la plus grande vraisemblance.


- Vérifier empiriquement comment évolue ce maximum de vraisemblance si l'on effectue de plus en plus de tirages
- Que se passe-t-il quand il y a trop de tirages ? Représenter la log-vraisemblance plutot que la vraisemblance dans ce cas.




```python
NbTirages=100
NbPiles = np.random.randn(NbTirages) > 0
NbPiles = np.sum(NbPiles)
NbFaces=NbTirages-NbPiles


PossibleValuesForP=np.linspace(0.01,0.99,100)

def vraisemblance(n_pile, n_face, p):
    return (p**n_pile)*((1-p)**n_face)

v = np.array([vraisemblance(n_pile=NbPiles, n_face=NbFaces, p=p)for p in PossibleValuesForP])

fig = plt.figure()
plt.plot(PossibleValuesForP, v, '-')
plt.title('Vraisemblance en fonction de p')
plt.show()
```

Quand on augmente le nombre de tirages, le maximum de vraisemblance tend vers 0 donc on atteint le 0 numérique.

```python
NbTirages= 100
NbPiles = np.random.randn(NbTirages) > 0
NbPiles = np.sum(NbPiles)
NbFaces=NbTirages-NbPiles

def log_vraisemblance(n_pile, n_face, p):
    return (n_pile*np.log(p)) + (n_face*np.log(1-p))

log_v = np.array([log_vraisemblance(n_pile=NbPiles, n_face=NbFaces, p=p) for p in PossibleValuesForP])

fig = plt.figure()
plt.plot(PossibleValuesForP, log_v, '-')
plt.title('Log-Vraisemblance en fonction de p')
plt.show()
```

```python
print(f"Maximum de vraisemblance : p={log_v.argmax()/len(PossibleValuesForP)}")
```
